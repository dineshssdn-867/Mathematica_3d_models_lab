# Undefined sinx and siny

The plot looks very bizarre and the reason it looks so strange is because when x is 0 or when y is 0 our function is undefined and so we can see that there's some very unusual behavior Mathematica is trying its best to plot but it it's having a hard time because of those places where the function is undefined

```
Plot3D[(Sin[x]+Sin[y])/(x * y), {x,-Pi,Pi}, {y, -Pi,Pi}]
```

# Saddle

A saddle point or minimax point is a point on the surface of the graph of a function where the slopes (derivatives) in orthogonal directions are all zero (a critical point), but which is not a local extremum of the function.

```
Plot3D[x^2-3y^2,{x,-5,5},{y,-5,5}]
```

```
a = 3; b = 2; c = 1;
 RegionPlot3D[
  x^2/a^2 + y^2/b^2 + z^2/c^2 < 1, {x, -3, 3}, {y, -2, 2}, {z, -1, 1},
   BoxRatios -> Automatic]
```

```
Clear[f,g,h,u,v,x,y,n,t]  (* Clear symbols used *)

f[x_,y_]=2Sin[2(x^2+y^2)];      (* Initial surface *)
g[x_,y_]=-x^2+y^2;                  (* Final surface *)
u[t_]=1-t;                  (* Disipation of initial surface *)
v[t_]=Sin[(5Pi*t)/2];                        (* Emergence of final surface *)
n=20;                              (* Number of steps to use *)

(* The full mapping function *)
h[x_,y_,t_]=u[t]*f[x,y]+v[t]*g[x,y];

(* Table of the surface plots showing the morphing *)
Table[Plot3D[h[x,y,i/n],{x,-2,2},{y,-2,2},AxesLabel->{"x","y","z"},AspectRatio->1,PlotPoints->40,Mesh->False,PlotRange->{-4,4}],{i,0,n}]
```
